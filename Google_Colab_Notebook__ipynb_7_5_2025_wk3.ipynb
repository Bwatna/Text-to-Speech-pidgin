```python
import pandas as pd
import string
import re
```

```python
df = pd.read_excel("Bible dataset in pidgin.xlsx")
df.head()
```

```python
print(df.isnull().sum())
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
def clean_text(text):
    text = text.lower()  # Normalize to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

df['clean_pidginstring'] = df['pidginstring'].apply(clean_text)
df[['pidginstring', 'clean_pidginstring']].head()
```

```python
df['word_count'] = df['clean_pidginstring'].apply(lambda x: len(x.split()))

df['sentence_count'] = 1

print("Total sentences:", df['sentence_count'].sum())
print("Average words per sentence:", df['word_count'].mean())
print("Total words:", df['word_count'].sum())

from collections import Counter
all_words = ' '.join(df['clean_pidginstring']).split()
word_freq = Counter(all_words)
print(word_freq.most_common(10))
```
